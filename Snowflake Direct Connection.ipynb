```
#define query
query = 'select * from db.transformed.customer limit 100'
```

```
#set up connection
user = dbutils.secrets.get(scope="", key="sf-p-user-001")
password = dbutils.secrets.get(scope="", key="sf")
database_host_url = 'https://snowflakecomputing.com'
#pull table
df = (spark.read
  .format("snowflake")
  .option("query", query)
  .option("sfUrl", database_host_url)
  .option("sfUser", user)
  .option("sfPassword", password)
  .option("sfDatabase", "")
  .option("sfSchema", "general_use_bas")
  .option("sfWarehouse", "")
  .load()
)
```

```
#Use this code to make a new database (also called schema) in the Data Catalog
spark.sql("USE CATALOG hive_metastore")
spark.sql("CREATE SCHEMA rgb_db")
```

```
#save table
df.write.saveAsTable('rgb_db.test_table')

```
