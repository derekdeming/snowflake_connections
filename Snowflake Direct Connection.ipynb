```
#define query
query = 'select * from db_swire_bi_p_edw.transformed.dim_mdm_customer_master limit 100'
```

```
#set up connection
user = dbutils.secrets.get(scope="azure_kv_p_001", key="sf-p-user-001")
password = dbutils.secrets.get(scope="azure_kv_p_001", key="sf-p-pwd-001")
database_host_url = 'https://swire.east-us-2.azure.snowflakecomputing.com'
#pull table
df = (spark.read
  .format("snowflake")
  .option("query", query)
  .option("sfUrl", database_host_url)
  .option("sfUser", user)
  .option("sfPassword", password)
  .option("sfDatabase", "db_bi_p_edw")
  .option("sfSchema", "general_use_bas")
  .option("sfWarehouse", "SWIRE_WH_ELDAMO_M")
  .load()
)
```

```
#Use this code to make a new database (also called schema) in the Data Catalog
spark.sql("USE CATALOG hive_metastore")
spark.sql("CREATE SCHEMA rgb_db")
```

```
#save table
df.write.saveAsTable('rgb_db.test_table')

```
